```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define an exponential cdf

```python
def exponential_distribution(λ, x):
    return 1.0 - λ*np.exp(-λ*x)
```

```python
λ = 2.0
```

# Generate exponential random numbers

```python
exp_rand = np.random.exponential(scale = 1/λ, size = 100000)
```

# Plot a histogram of exponential random numbers

```python
plt.hist(exp_rand)
```

# Calculate F(X) by plug in exponential random numbers into the function exponential_distribution

```python
unif_rand = exponential_distribution(2,exp_rand)
```

# Plot a histogram of those numbers to see the distribution of them numerically

```python
plt.hist(unif_rand)
```
